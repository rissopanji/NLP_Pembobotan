## 1. Import Library

```python
import pandas as pd
import re
from docx import Document
import nltk
from nltk.tokenize import word_tokenize
import numpy as np
```

## 2. Load Dataset

```python
data = Document('Berita - httpswww.bbc.comindonesiaarticlescprgdg2zl3eo.docx')

for para in data.paragraphs:
    print(para.text)
```

## 3. Preposessing

Casefolding

```python
def case_fold(text):
    return text.lower()

for para in data.paragraphs:
    text = para.text
    text = case_fold(text)
    print(text)
```

Filtering Tanda Baca

```python
def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

for para in data.paragraphs:
    text = para.text
    text = case_fold(text)
    text = remove_punctuation(text)
    print(text)
```

Filtering Stoplist 

(https://www.kaggle.com/datasets/oswinrh/indonesian-stoplist?resource=download)

```python
stopwords_file = 'stopwordbahasa.csv'

# Fungsi untuk membaca kamus stop words
def load_stopwords(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        stopwords = [line.strip() for line in f]
    return set(stopwords)

# Fungsi untuk menghapus stop words dari teks
def remove_stopwords(text, stopwords):
    tokens = text.split()
    filtered_tokens = [word for word in tokens if word.lower() not in stopwords]
    return ' '.join(filtered_tokens)

# Load stop words
stopwords = load_stopwords(stopwords_file)

for para in data.paragraphs:
    text = para.text
    text = case_fold(text)
    text = remove_punctuation(text)
    text = remove_stopwords(text, stopwords)
    print(text)
```

Stemming porter (bahasa Indonesia)

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

factory = StemmerFactory()
stemmer = factory.create_stemmer()

def perform_stemming(text, stemmer):
    return stemmer.stem(text)

for para in data.paragraphs:
    text = para.text
    text = case_fold(text)
    text = remove_punctuation(text)
    text = remove_stopwords(text, stopwords)
    text = perform_stemming(text, stemmer)
    print(text)
```

Simpan Data

```python
# Dokumen baru untuk menyimpan hasil
doc_clean = Document()

for para in data.paragraphs:
    text = para.text
    text = case_fold(text)
    text = remove_punctuation(text)
    text = remove_stopwords(text, stopwords)
    text = perform_stemming(text, stemmer)
    doc_clean.add_paragraph(text)

# Menyimpan hasil tokenisasi ke dalam DataFrame
doc_clean.save('dokumen_bersih.docx')
```

Tokenisasi

```python
# Load data
data = Document('dokumen_bersih.docx')

# Gabungkan semua teks dalam dokumen menjadi satu string
text = '\n'.join([para.text for para in data.paragraphs])

# Tokenisasi
tokens = nltk.word_tokenize(text)

# Buat dokumen baru
doc = Document()

for token in tokens:
    doc.add_paragraph(token)

# Simpan dokumen
doc.save('dokumen_bersih_tokenisasi.docx')
```

## 4. Pembobotan (TF-IDF)

Ubah tipe data list

```python
def doc_to_list(doc):
    word_list = []
    for paragraph in doc.paragraphs:
        for word in paragraph.text.split():
            word_list.append(word)
    return word_list

# Baca dokumen
doc = Document("dokumen_bersih_tokenisasi.docx")

# Konversi ke list
dokumen_token = doc_to_list(doc)

print(dokumen_token)
```

Term Frequency (TF)

```python
def calc_TF(word_list):

    word_count = {}
    total_words = len(word_list)
    for word in word_list:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
    

    TF_dict = {}
    for word, count in word_count.items():
        TF_dict[word] = count / total_words
    
    return TF_dict

def print_TF(tf_dict):
    for word, tf in tf_dict.items():
        print(f"'{word}': {tf}")


dokumen_list = dokumen_token

TF = calc_TF(dokumen_list)

print_TF(TF)
```

Inverse Document Frequency (IDF)

```python
def calc_DF(tfDict):
    count_DF = {}

    for document in tfDict:
        word_count = len(set(document.split()))
        for term in document.split():
            if term in count_DF:
                count_DF[term] += 1 / word_count 
            else:
                count_DF[term] = 1 / word_count 
    return count_DF

def print_TF(tfDict):
    for term, freq in tfDict.items():
        print(f"{term} ; {freq}")


DF = calc_DF(dokumen_list)

print_TF(DF)
```

```python
n_document = len(dokumen_list)

def calc_IDF(__n_document, __DF):
    IDF_Dict = {}
    for term in __DF:
        IDF_Dict[term] = np.log(__n_document / (__DF[term] + 1))
    return IDF_Dict
  

IDF = calc_IDF(n_document, DF)

print_TF(IDF)
```

TF-IDF

```python
def calc_TF_IDF(TF, IDF):
    TF_IDF_Dict = {}

    for key in TF:
        TF_IDF_Dict[key] = TF[key] * IDF[key]
    return TF_IDF_Dict


TF_IDF = calc_TF_IDF(TF, IDF)
print_TF(TF_IDF)

```

Sparse Matrix

```python
sorted_DF = sorted(DF.items(), key=lambda kv: kv[1], reverse=True)[:50]

unique_term = [item[0] for item in sorted_DF]

def calc_TF_IDF_Vec(__TF_IDF_Dict):
    TF_IDF_vector = [0.0] * len(unique_term)

    for i, term in enumerate(unique_term):
        if term in __TF_IDF_Dict:
            TF_IDF_vector[i] = __TF_IDF_Dict[term]
    return TF_IDF_vector


TF_IDF_Vec = calc_TF_IDF_Vec(TF_IDF)


print("First row matrix TF_IDF_Vec Series\n")
print(TF_IDF_Vec)


print("\nmatrix size : ", len(TF_IDF_Vec))
```
